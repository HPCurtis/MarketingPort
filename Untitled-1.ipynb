```python
data_path = "https://raw.githubusercontent.com/matheusfacure/python-causality-handbook/master/causal-inference-for-the-brave-and-true/data/smoking.csv"

raw_data_df = pd.read_csv(data_path)

raw_data_df.head()

# select relevant columns
df = raw_data_df.copy().drop(columns=["lnincome", "beer", "age15to24"])

features = ["cigsale", "retprice"]

pre_df = (
    df
    .query("~ after_treatment")
    .pivot(index='state', columns="year", values=features)
    .T
)

post_df = (
    df
    .query("after_treatment")
    .pivot(index='state', columns="year", values=features)
    .T
)

idx = 3

y_pre = pre_df[idx].to_numpy()
x_pre = pre_df.drop(columns=idx).to_numpy()
pre_years = pre_df.reset_index(inplace=False).year.unique()
n_pre = pre_years.size

y_post = post_df[idx].to_numpy()
x_post = post_df.drop(columns=idx).to_numpy()
post_years = post_df.reset_index(inplace=False).year.unique()
n_post = post_years.size

k = pre_df.shape[1] - 1
```

```python
stanSynthticControl = ''' 
data{
int N;
int K;
array[N] real y;
matrix[N,K] X;

// Priro value for sd of sigma parameter
real sdS;

}
parameters{
simplex[K] beta;
real<lower = 0> sigma; 
}

transformed parameters{
vector[N] mu;
mu = X * beta;
}

model{
// Priors 
sigma ~ normal(0, sdS);
// Uniform default prior
beta ~ dirichlet( 1.0/K * rep_vector(1, K) );

// Likelihood
y ~ normal(mu, sigma);
}
generated quantities{
// Posterior predictive checksS
real yrep[N];
yrep = normal_rng(mu, sigma);
}
'''
```

```python
data = {'N': len(pre_df),
        'K': k,
        'y': y_pre,
        'X':x_pre,
        'sdS': 5
        }

sm = ps.build(stanSynthticControl,data= data)
```

```python
fit = sm.sample()
```

```python
az.summary(fit)
```

```python
az.plot_forest(data=fit, combined=True, var_names=["beta"]);
```

```python
# Convert the pystan object into Arviz inference object for use in plotting functions.
idata = az.from_pystan(posterior=fit,posterior_model=sm, posterior_predictive=['yrep'],observed_data= 'y') 
                

az.plot_ppc(data = idata, data_pairs = {'y': 'yrep'}, num_pp_samples = 500);
```
