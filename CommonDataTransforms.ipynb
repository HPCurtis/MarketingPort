```python
import numpy as  np 
import pandas as  pd
import matplotlib.pyplot as plt
import scipy.stats
```

The pearson correlation is desinged to analyse random variables that show linear and 
normal distributions.


```python
# We can generate data that has stong non-linear association.
x = np.random.uniform(-100, 100, size = 1000)
x2 = x ** 2
# Plot data with non-linear relationship.
plt.scatter(x = x, y = x2)

# But get near zero correlation coefficents
print(np.corrcoef(x, x2))
```

```python
df = pd.read_csv('http://goo.gl/PmPkaG')

distanceStore = df['distance.to.store']
distInverse = 1/df['distance.to.store']
distSInverse = 1/np.sqrt(df['distance.to.store'])

print('Pearson Correlation between the distance from store and and amount of money spent at the store', scipy.stats.pearsonr(distanceStore, df['store.spend'])[0])
print('Pearson Correlation between the the inverse distance (1/distance from the store) from store and and amount of money spent at the store', scipy.stats.pearsonr(distInverse, df['store.spend'])[0])
print('Pearson Correlation between the the inverse square root distance (1/sqrt(distance) from the store) from store and and amount of money spent at the store', scipy.stats.pearsonr(distSInverse, df['store.spend'])[0])
```

```python
fig, axs = plt.subplots(1, 3,  figsize=(25, 5)) 

axs[0].scatter(x = distanceStore, y = df['store.spend'], c='black', alpha=.3)
axs[0].set_xlabel('Distance to store')
axs[1].scatter(x = distInverse, y = df['store.spend'], c='black', alpha=.3)
axs[1].set_xlabel('Inverse distance to store')
axs[2].scatter(x = distSInverse, y = df['store.spend'], c='black', alpha=.3)
axs[2].set_xlabel('Inverse sqaure root distance to store')
```
